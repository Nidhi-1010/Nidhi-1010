```python
exchange_rates = {
    'USD': 1.0,  # USD to USD rate is 1
    'EUR': 1.12,  # Example: 1 USD = 1.12 EUR
    'GBP': 1.31,  # Example: 1 USD = 1.31 GBP
    'AUD': 0.65,
    'BRL': 0.20,
    'CAD' : 0.74,
    'CHF': 1.15,
    'CLP': 0.0010,
    'CNY': 0.14,
    'DKK':0.14,
    'HUF': 0.0028,
    'INR': 0.012,
    'JPY': 0.0067,
    'MXN':0.058,
    'PLN':0.25,
    'SGD': 0.74,
    'TRY':0.033
}

# Convert salary to USD based on exchange rates and create a new column
data['salary_in_usd'] = data.apply(lambda row: row['salary'] * exchange_rates.get(row['salary_currency'], 1.0), axis=1)

# Print the DataFrame to verify the new column
print(data)
```

```python
# Calculate mean salary per job title
mean_salary_per_job_title = data.groupby('job_title')['salary_in_usd'].mean()

# Sort by mean salary in descending order
mean_salary_per_job_title = mean_salary_per_job_title.sort_values(ascending=False)

# Format the output
mean_salary_per_job_title = mean_salary_per_job_title.apply(lambda x: "{:.2f}".format(x))
mean_salary_per_job_title
```

```python
#Calculate mean salary per experience level
mean_salary_per_experience_level = data.groupby('experience_level')['salary_in_usd'].mean()
# Sort by mean salary in ascending order
mean_salary_per_experience_level = mean_salary_per_experience_level.sort_values(ascending=True)

# Format the output
mean_salary_per_experience_level =mean_salary_per_experience_level .apply(lambda x: "{:.2f}".format(x))
mean_salary_per_experience_level

# Prepare data for plotting
experience_level = mean_salary_per_experience_level.index.to_list()

mean_salaries = mean_salary_per_experience_level.values

# Create and configure the bar chart
plt.figure(figsize=(10, 10))
plt.bar(experience_level, mean_salary_per_experience_level,color="red")

# Add labels and title
plt.xlabel("Experince Level")
plt.ylabel("Mean Salary")
plt.title("Mean Salary by Experience Level")

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)

# Show the plotin t
plt.tight_layout()
plt.show()
```
