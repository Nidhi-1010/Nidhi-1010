# **Feature Selection**

```python
data.sample(5)
```

```python
data['quality'].unique()
```

```python
#If wine quality is gretaer than 7 then we will consider it as good quality data otherwise bad quality
data['good_quality']=[1 if x>=7 else 0 for x in data['quality']]
data.sample(5)
```

```python
data['good_quality'].value_counts()
```

```python
#Seperate dependent and independent variables
X=data.drop(['quality','good_quality'],axis=1)
Y=data['good_quality']
```

```python
#Correaltion Between Variables
data.corr()
```

```python
#Heat Map
plt.figure(figsize=(10,6))
sns.heatmap(data.corr(),cmap='coolwarm')
```

```python
#SETTING INDEPENDENT AND DEPENDENT VARIABLES
Y = data['quality'].apply(lambda y: 1 if y>7 else 0)
Y
```

```python

x = data.drop('quality',axis=1)
```

```python
#TRAIN AND TEST SPLIT

from sklearn.model_selection import train_test_split

x_train, x_test, Y_train, Y_test = train_test_split(x,Y,test_size= 0.2,random_state = 0)

from sklearn.preprocessing import StandardScaler


stdscale = StandardScaler().fit(x_train)
x_train_std = stdscale.transform(x_train)
x_test_std = stdscale.transform(x_test)
```

```python
#MODEL TRAINING

#Logistic Regression

from sklearn.linear_model import LogisticRegression


logreg = LogisticRegression()
logreg.fit(x_train_std, Y_train)

LogisticRegression()
```

```python


y_pred = logreg.predict(x_test_std)
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


accuracy_score(Y_test, y_pred)
```

```python

print(classification_report(Y_test, y_pred))

```

```python
print(confusion_matrix(Y_test, y_pred))
```

```python
#Rabdom Forest Classifier

from sklearn.ensemble import RandomForestClassifier


rfc = RandomForestClassifier()
rfc.fit(x_train_std, Y_train)

rfc
```

```python
rfc_pred = rfc.predict(x_test_std)


accuracy_score(Y_test, rfc_pred)
```

```python
rfc.feature_importances_
```

```python
pd.Series(rfc.feature_importances_, index= data.drop('quality',axis=1).columns).plot(kind='barh')
```

```python
#Decision Tree classifier

from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train_std, Y_train)
accuracy_score(Y_test, dt.predict(x_test_std))
```
